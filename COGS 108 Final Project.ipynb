# The Effects of US Political Leanings on the Number of Covid-19 Cases

# Overview

Our research explores whether a particular US region’s overall political affiliation affects their total number of Covid-19 cases, exploring additional factors such as education ranking and Covid-19 safety guidelines. Using linear regression and multiple linear regression, we explored the relationship between these variables and found that there is a correlation between the political leaning and the safety policies of a US region and its overall amount of Covid cases.

# Research Question

How does the political affiliation of a particular US State impact the number of Coronavirus cases present in that state? Does a state’s overall educational level and adherence to CDC social distancing measures influence the residents’ inclination to follow such policies and thus the number of positive cases? Furthermore, within California and within Mississippi, how do two counties with opposing political leanings differ in the number of COVID-19 cases?

# Hypothesis


We hypothesize that if a particular state is more politically aligned with Republican values, then they may be more inclined to disregard CDC guidelines and therefore have a higher number of Covid-19 cases. We similarly hypothesize that a state with a higher overall education level and stricter adherence to Covid-19 safety measures will have a lower total number of positive cases. Because California is mainly Democratic and Mississippi is mainly Republican, we hypothesize that politically contrasting counties within these states will follow the overall trend of their state in terms of following safety measures.

## State Social Distancing Policies (Dataset 10)

```python
policies = pd.read_csv("Data/state_social_distancing_actions.csv")
policies.head()
```

```python
#Clean Up Column Names 
policies.columns = ['State', 'Reopening', 'Stay at Home', 'Traveling Mandatory Quarantine', 'Business Closures', 'Gathering Ban',
                      'Restaurant Limits', 'Bar Closures', 'Mask Requirements', 'State of Emergency']
#Drop any rows that are not states
policies = policies.iloc[2: , :]
policies = policies.reset_index(drop=True)
policies.head()
```

Here we first read state_social_distancing_actions.csv into the policies dataframe. Next, we dropped the columns we did not need, and renamed the other columns for maximum readability. Then we dropped the rows of the dataframe that did not have necessary information about the states and reset the indices. 

```python
policies['Reopening'].unique()
```

```python
policies['Reopening'] = policies['Reopening'].astype(str)
def standardize_reopening(string):
    string = string.replace('Proceeding with Reopening', 'Reopening')
    string = string.replace('New Restrictions Imposed', 'New Restrictions')
    return string 

policies['Reopening'] = policies['Reopening'].apply(standardize_reopening)
policies['Reopening'].unique()
```

```python
policies['Stay at Home'].unique()
```

```python
policies['Stay at Home'] = policies['Stay at Home'].astype(str)
def standardize_stayHome(string):
    string = string.replace('-', 'nan')
    string = string.replace('Rolled Back to High Risk Groups', 'High Risk Stay Home')  
    string = string.replace('Statewide (New)', 'Statewide')
    return string 

policies['Stay at Home'] = policies['Stay at Home'].apply(standardize_stayHome)
policies['Stay at Home'].unique()
```

```python
policies['Traveling Mandatory Quarantine'].unique()
```

```python
policies['Traveling Mandatory Quarantine'] = policies['Traveling Mandatory Quarantine'].astype(str)
def standardize_travel(string):
    string = string.replace('-', 'nan')
    string = string.replace('All Travelers (New)', 'All Travelers')    
    string = string.replace('From Certain States (New)', 'From Certain States')    
    string = string.replace('Rolled Back to International Travel', 'No International Travel') 
    string = string.replace('Rolled Back to Certain States', 'From Certain States') 
    return string 

policies['Traveling Mandatory Quarantine'] = policies['Traveling Mandatory Quarantine'].apply(standardize_travel)
policies['Traveling Mandatory Quarantine'].unique()
```

```python
policies['Business Closures'].unique()
```

```python
policies['Business Closures'] = policies['Business Closures'].astype(str)
def standardize_businesses(string):
    string = string.strip()
    string = string.replace("All Non-Essential Businesses Permitted to Reopen with Reduced Capacity", "Reopen Non Essential Businesses; Reduced Capacity")    
    string = string.replace("All Non-Essential Businesses Permitted to Reopen", "Reopen Non Essential Businesses")    
    string = string.replace("New Business Closures or Limits", "Close Business") 
    string = string.replace("Some Non-Essential Businesses Permitted to Reopen with Reduced Capacity", "Reopen Non Essential Businesses; Reduced Capacity") 
    string = string.replace("Some Non-Essential Businesses Permitted to Reopen", "Reopen Non Essential Businesses") 
    string = string.replace('-', 'nan')
    return string 

policies['Business Closures'] = policies['Business Closures'].apply(standardize_businesses)
policies['Business Closures'].unique()
```

```python
policies['Gathering Ban'].unique()
```

```python
policies['Gathering Ban'] = policies['Gathering Ban'].astype(str)
def standardize_gatherings(string):
    string = string.replace('New Limit on Large Gatherings in Place', 'Limit Large Gatherings')    
    string = string.replace('>50 Prohibited', '50 or Less')
    string = string.replace('All Gatherings Prohibited', 'No Gathering') 
    string = string.replace('>10 Prohibited', '10 or Less')
    string = string.replace('>25 Prohibited', '25 or Less')
    return string 

policies['Gathering Ban'] = policies['Gathering Ban'].apply(standardize_gatherings)
policies['Gathering Ban'].unique()
```

```python
policies['Restaurant Limits'].unique()
```

```python
policies['Restaurant Limits'] = policies['Restaurant Limits'].astype(str)
def standardize_restaurants(string):
    string = string.strip()
    string = string.replace('Reopened to Dine-in Service with Capacity Limits', 'Dine In Allowed; Limited Capacity') 
    string = string.replace('Reopened to Dine-in Service', 'Dine In Is Allowed')    
    string = string.replace('New Service Limits', 'Service Limits')    
    string = string.replace('Newly Closed to Indoor Dining', 'No Indoor Dining') 
    string = string.replace('-','nan')
    return string 

policies['Restaurant Limits'] = policies['Restaurant Limits'].apply(standardize_restaurants)
policies['Restaurant Limits'].unique()
```

```python
policies['Bar Closures'] = policies['Bar Closures'].astype(str)
def standardize_bars(string):
    string = string.replace('Newly Closed', 'Closed')    
    string = string.replace('New Service Limits', 'Service Limits')
    return string 

policies['Bar Closures'] = policies['Bar Closures'].apply(standardize_bars)
policies['Bar Closures'].unique()
```

```python
policies['Mask Requirements'] = policies['Mask Requirements'].astype(str)
def standardize_masks(string):
    string = string.replace('\n', ' ')
    string = string.replace('Required for General Public', 'Required for All')    
    string = string.replace('Required for Certain Employees', 'Required for Employees')
    string = string.replace('Required for Certain Employees;\nAllows Local Officials to Require for General Public', 'Required for Employees, Local Officials Choose for Public')
    string = string.replace('Allows Local Officials to Require for General Public', 'Local Officials Choose for Public')
    string = string.replace('-','nan')
    return string 

policies['Mask Requirements'] = policies['Mask Requirements'].apply(standardize_masks)
policies['Mask Requirements'].unique()
```

```python
def quantify_Reopening(string):
    #state of reopening
    string = string.replace('nan', '0')
    string = string.replace('Reopened', '1')
    string = string.replace('Reopening', '2')
    string = string.replace('New Restrictions', '3')
    string = string.replace('Paused', '4')
    return int(string)

def quantify_StayAtHome(string):
    #stay at home
    string = string.replace('nan', '0')
    string = string.replace('Lifted', '1')
    string = string.replace('High Risk Stay Home', '2')
    string = string.replace('Statewide', '3')
    return int(string)

def quantify_Traveling(string):
    #traveling
    string = string.replace('nan', '0')
    string = string.replace('Lifted', '1')
    string = string.replace('No International Travel', '2')
    string = string.replace('From Certain States', '2')
    string = string.replace('All Travelers', '3')
    return int(string)
    
def quantify_Business(string):
    #business closures 
    string = string.replace('nan', '0')
    string = string.replace('Reopen Non Essential Businesses; Reduced Capacity', '2')
    string = string.replace('Reopen Non Essential Businesses', '1')
    string = string.replace('Close Business', '3')
    return int(string)

def quantify_Gathering(string):
    #gathering ban
    string = string.replace('No Limit', '1')
    string = string.replace('50 or Less', '2')
    string = string.replace('25 or Less', '3')
    string = string.replace('10 or Less', '4')
    string = string.replace('No Gathering', '5')
    return int(string)

def quantify_Restaurants(string):
    #restaurant limits
    string = string.replace('nan', '0')
    string = string.replace('Dine In Is Allowed', '1')
    string = string.replace('Dine In Allowed; Limited Capacity', '2')
    string = string.replace('Service Limits', '3')
    string = string.replace('No Indoor Dining', '4')
    return int(string)
    
def quantify_Bars(string):
    #bar closures
    string = string.replace('Reopened', '1')
    string = string.replace('Service Limits', '2')
    string = string.replace('Closed', '3')
    return int(string)
    
def quantify_Masks(string):
    #mask requirements
    string = string.replace('nan', '0')
    string = string.replace('Required for Employees; Local Officials Choose for Public', '3')
    string = string.replace('Local Officials Choose for Public', '1')
    string = string.replace('Required for Employees', '2')
    string = string.replace('Required for All', '4')
    return int(string)
   
```

### Political Rankings at US State and County levels 

```python
def political_rank(difference):
    # neutral -- only a 5% difference 
    if difference <= 5 and difference >= -5:
        rank = 3
    # 10% difference -- positive, mildly Democratic 
    elif difference > 5 and difference <= 10:
        rank = 2
    # >10% difference -- positive, mainly Democratic 
    elif difference > 10:
        rank = 1
    # 10% difference -- negative, mildly Republican 
    elif difference < -5 and difference >= -10:
        rank = 4
    # >-10% difference -- negative, mainly Republican 
    elif difference < -10:
        rank = 5
    return rank
```

The above cell is a defined function that creates a political rank for each state based on the percentage difference column of the dataframe. The percentage difference is calculated by subtracting the 'Trump % votes' column from the 'Biden % votes' column. Then we defined a 5 level scale to determine what the individual states' political leanings were based off of the 2020 US Election. If only a difference between positive and negative 5% existed between the two candidates, the state was ranked a 3. If a positive difference between 5 and 10 existed, then the state was deemed mildly democratic and ranked a 2. If a positive percentage of more than 10 existed, then the state was considered primarily democratic and ranked as a 1. If the percentage difference fell between -5% and -10% the state was seen as mildly republican and ranked at a 4. Finally if the percentage difference fell past -10% the state was considered primarily republican and ranked as a 5. 

```python
US_data["% Difference"] = US_data["Biden % votes"] - US_data["Trump % votes"]
US_data['PoliticalRank'] = US_data['% Difference'].apply(political_rank)
US_data.head()
```

Here the percentage difference between 'Biden % votes' and 'Trump % votes' was created for every US state, and using that and the defined function above a rank column was created to hold the political ranking of each state.

Just like we did with the individual states, we will create a Linear Regression Model predicting the amount of COVID cases using Political Rank. However, this time we are analyzing Mississippi's individual counties because California is a primarily Republican State. We want to see if there is a difference between California and Mississippi.

```python
MS_outcome, MS_predictors = patsy.dmatrices('PercentCases ~ PoliticalRank', MS_data)

MS_model = sm.OLS(MS_outcome, MS_predictors)
MS_results = MS_model.fit()
```

```python
print(MS_results.summary())
```

### Analyzing our MS Counties Linear Regression model

Looking at our results we obtain the following model:
$$ PercentCases = 6.3803 - 0.2269 * Political Rank$$

Here for the Mississippi Counties the model seems to predict that the % of COVID cases decreases by 0.22% with an increase in Political Rank according to the data. However, looking at the data the intercept for Mississippi is much higher compared to California (6.38 compared to 2.70). The large intercept indicates that the average infection rate for a County in Mississippi is larger than California's average. The R-squared value is 0.1 which is extremely low even assuming that people are unpredictable. 

Considering the fact that Mississippi is a Republican State, the fact that it has a much higher intercept for Percentage of Covid Cases could show a correlation between Republican political policies on prevention and rate of COVID Infections.
